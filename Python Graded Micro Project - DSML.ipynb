```python
# Q.4) In a single terror incident in India, find out top 5 cities by number killed

c = df1[(df1['country_txt']=='India') & (df1['nkill']>=1)]
c = c.sort_values(by='nkill', ascending=False)
d = c[['country_txt', 'city', 'nkill']]
d.head(5)
        
```

```python
# Q.5) In a single terror incident in India, find out top 5 cities by number killed and wounded
india_attacks = df1[df1['country_txt']=='India']
city_kills = india_attacks.sort_values(by=['nkill', 'nwound'], ascending=False)
cities = city_kills[['country_txt', 'city', 'nkill', 'nwound']]
cities.head(5)
```

#### Top 5 cities are Mumbai, Jhargam, Bombay, Dantewada district, Banabari

```python
#Q.6) How many attacks were successful that were suicide attacks?
c=df1[(df1['success']==1)&(df1['suicide']==1)].shape[0]
print(c, 'is the count of attacks that were successful, that were suicidal attacks')
```

```python
#Q.7) Label all the incidents where the number killed was more than 5 as severe.
def severity(x):
    if x>5:
        return "severe"
    else:
        return "not severe"
    
```

```python
df1['Severity'] = df1['nkill'].map(severity)
df1.head()
```

```python
#Q.8) write a function to label an incident that was both successful and suicidal

def success_suicide(x):
    if x['success']==1 & x['suicide']==1:
        return "The incident is both successful and suicidal"
    else:
        return "The incident is not successful and suicidal"
    
```

```python
df1['Success and Suicide'] = df1.apply(success_suicide, axis=1)
df1.head()
```

```python
#Q.9) Create a new category representing if the incident occured in Afghanistan,Pakistan or India as one level 
#of the category and all the other countries as
#another level. Label all other countries as ROW and new column which contains
#the new category as â€˜Localâ€™ in the data.

df1['Local'] = df1['country_txt'].apply(lambda x: 'Af-Pak-Ind' if x in ['Afghanistan', 'Pakistan', 'India'] else 'ROW')
df1.head()
```

```python
# Q.10) How many incidents happened in Af-Pak-India vs ROW?
df1['Local'].value_counts()
```

```python
# Q.10) How many incidents happened in Af-Pak-India vs ROW?
print('Incidents happened in ROW is:',df1['Local'].value_counts()['ROW'])
print('Incidents happened in Af-Pak-Ind is:',df1['Local'].value_counts()['Af-Pak-Ind'])
```
