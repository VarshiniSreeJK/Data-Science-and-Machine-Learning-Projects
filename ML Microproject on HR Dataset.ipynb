### Handling Zero Values in NumCompaniesWorked

- The minimum value for "NumCompaniesWorked" is zero, which is not realistic if a person's total experience is greater than 1. Let's replace these zero values with the mode.

```python
zero = df[df['NumCompaniesWorked'] == 0].index
mode_value = df['NumCompaniesWorked'].mode()[0]

df.loc[zero, 'NumCompaniesWorked'] = mode_value
```

### Performing Log transformation for positive skewness on MonthlyIncome


```python
#checking if monthly Income has 0 and positive values

postive_value = all(value > 0 for value in df['MonthlyIncome'])


zero_income = 0 in df['MonthlyIncome']


print( postive_value)
print(zero_income)
```

```python
df['MonthlyIncomeLog'] = np.log1p(df['MonthlyIncome'])
```

#### Inference : 

1. MonthlyIncome has right-skewed distributions and inorder to make it more symmetric Log transformation is used

2. MonthlyIncome column has both zero and positive values hence np.log1p function is used.

3. The np.log1p function is used to handle zero or very small values in the data. It adds 1 to each value before taking the logarithm, preventing issues with the logarithm of zero.

### Square Root transformation 

- Used to reduce the impact of extreme values

```python
df['TotalWorkingYearsLog'] = np.sqrt(df['TotalWorkingYears'])
```
